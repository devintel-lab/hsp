```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df202 = pd.read_csv("../script_utils/dist_change_exp202.csv")
```

```python
df204 = pd.read_csv("../script_utils/dist_change_exp204.csv")
```

```python
g = sns.FacetGrid(df204, col="target", sharex=False, col_wrap=4)
g.map_dataframe(sns.scatterplot, x="trial_num", y="kl_target", color='red')
g.map_dataframe(sns.scatterplot, x="trial_num", y="kl_prev", color='blue')
g.map_dataframe(sns.scatterplot, x="trial_num", y="ent_p", color='green')
g.set_axis_labels("Trial Num", "KL Distance")
g.add_legend()
plt.savefig("exp204_kl_distances.png", dpi=300)
```

```python
g = sns.FacetGrid(df202.query('block_set == 1'), col="target", sharex=False, col_wrap=4)
g.map_dataframe(sns.scatterplot, x="trial_num", y="kl_target", color='red')
g.map_dataframe(sns.scatterplot, x="trial_num", y="kl_prev", color='blue')
g.map_dataframe(sns.scatterplot, x="trial_num", y="ent_p", color='green')
g.set_axis_labels("Trial Num", "KL Distance")
g.add_legend()
plt.savefig("exp202_kl_distances_blockset1.png", dpi=300)
```
